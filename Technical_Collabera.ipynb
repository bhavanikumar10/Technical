```python
# Task B

# loading the first csv data file
A_df = pd.read_csv("TechnicalScreenB_collab.csv")
A_df.head()
```

```python
# loading the second csv data file
B_df = pd.read_csv("TechnicalScreenc_collab.csv")
B_df.head()
```

```python
# Merging the two dataframes

merge_df = pd.merge(A_df, B_df, on="Account.Number", how = "left")
merge_df
```

```python
# Summary of Productwise volume

groupedproduct_df = merge_df.groupby('Product')
groupedproduct_df.sum()
```

```python
# calculate percent difference between product volumes in 
# 2020 and 2021.

merge_df["Date"] = pd.to_datetime(merge_df["Date"])

merge_df['year'] = merge_df["Date"].dt.year

merge_df.head()
```

```python
groupedyear1_df = merge_df.groupby(['Product', 'year'])
newagg_df1 = groupedyear1_df.sum()
newagg_df1
```

```python
groupedyear_df = merge_df.groupby(['Product', 'year'])
newagg_df = groupedyear_df.sum()
newagg_df['% change'] = newagg_df.pct_change()
newagg_df

```

```python
# Countrywise defect quantities

countrygroup = merge_df.groupby("Country")
csum = countrygroup.sum()
csum
```

```python
# sorting countries on the their defect volumes
csort = csum.sort_values(by = "Quantity")
csort
```

```python
# top 3 countries with the highest complaint volume

top3 = csort.head(3)
top3
```
